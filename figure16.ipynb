```python
import pandas as pd
import matplotlib.pyplot as plt

data_source1 = pd.read_excel('figure16-1.xlsx')
# plot()
#print(data_source['Susceptible(EB)'])
print(data_source1)
figsize = 5,4
figure, ax = plt.subplots(figsize=figsize)
plt.plot(data_source1['Susceptible(group1)'],'-r',label='Susceptible(group1)',linewidth=2.0,ms=1)
plt.plot(data_source1['Exposed(group1)'],'b-',label='Exposed(group1)',linewidth=2.0,ms=1)
plt.plot(data_source1['Symptomatic(group1)'],'black',label='Symptomatic(group1)',linewidth=2.0,ms=1)
plt.plot(data_source1['Asymptomatic(group1)'],'gray',label='Asymptomatic(group1)',linewidth=2.0,ms=1)
plt.plot(data_source1['Intreated(group1)'],'green',label='Intreated(group1)',linewidth=2.0,ms=1)
plt.plot(data_source1['Recovered(group1)'],'gold',label='Recovered(group1)',linewidth=2.0,ms=1)
plt.plot(data_source1['Dead(group1)'],'pink',label='Dead(group1)',linewidth=2.0,ms=1)
plt.axis([0,120,0,21000])
# x
font1 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 10,
}
plt.title('Simulation Results of Group 1',font1)
plt.grid(linestyle="--")  # 
#
plt.legend(loc=5, numpoints=1,prop=font1)
plt.xlabel("Days",font1)
# y
plt.ylabel("The Number of Different Type Persons",font1)
#
plt.tick_params(labelsize=10)
labels = ax.get_xticklabels() + ax.get_yticklabels()
[label.set_fontname('Times New Roman') for label in labels]
plt.savefig('F:// //ABS///figure16-1.png',bbox_inches='tight')
```
