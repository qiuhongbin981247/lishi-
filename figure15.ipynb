```python
import pandas as pd
import matplotlib.pyplot as plt

data_source = pd.read_excel('figure15.xlsx')
# 函数plot()尝试根据数字绘制出有意义的图形
#print(data_source['Susceptible(EB)'])
print(data_source)
```

```python
figsize = 5,4
figure, ax = plt.subplots(figsize=figsize)
plt.plot(data_source['Susceptible(CTM)'],'-r',label='Susceptible(CTM)',linewidth=2.0,ms=1)
plt.plot(data_source['Susceptible(long-term LM)'],'b-',label='Susceptible(long-term LM)',linewidth=2.0,ms=1)
plt.plot(data_source['Susceptible(short-term LM)'],'green',label='Susceptible(short-term LM)',linewidth=2.0,ms=1,linestyle='-.')
plt.axis([0,75,0,20000])
# 设置x轴名称
font1 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 10,
}
plt.title('Simulation Results of the Susceptible In Different Strategies',font1)
plt.grid(linestyle="--")  # 设置背景网格线为虚线
#设置图例
plt.legend(loc=1, numpoints=1,prop=font1)
plt.xlabel("Days",font1)
# 设置y轴名称
plt.ylabel("The Number of the Susceptible",font1)
#保存图片
plt.tick_params(labelsize=10)
labels = ax.get_xticklabels() + ax.get_yticklabels()
[label.set_fontname('Times New Roman') for label in labels]
plt.savefig('F:/要多写多发期刊论文啊/社区防疫 论文/做工大的吧/纯ABS要能跑出来我去死/格式冲冲冲/图重新绘制/figure15-1.png',bbox_inches='tight')
```
