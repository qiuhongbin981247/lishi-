```python
import pandas as pd
import matplotlib.pyplot as plt

data_source = pd.read_excel('figure10.xlsx')
# 函数plot()尝试根据数字绘制出有意义的图形
#print(data_source['Susceptible(EB)'])
print(data_source)
```

```python
figsize = 5,4
figure, ax = plt.subplots(figsize=figsize)
plt.plot(data_source['S(Taking No Measure)'],'-r',label='Susceptible(NOCTM)',linewidth=2.0,ms=1)
plt.plot(data_source['S(Taking CT Measure)'],'b-',label='Susceptible(CTM)',linewidth=2.0,ms=1)
plt.axis([0,100,0,20000])
# 设置x轴名称
font1 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 10,
}
plt.title('Comparison With Taking CTM About the Susceptible Number',font1)
plt.grid(linestyle="--")  # 设置背景网格线为虚线
#设置图例
plt.legend(loc=0, numpoints=1,prop=font1)
plt.xlabel("Days",font1)
# 设置y轴名称
plt.ylabel("The Susceptible Number",font1)
#保存图片
plt.tick_params(labelsize=10)
labels = ax.get_xticklabels() + ax.get_yticklabels()
[label.set_fontname('Times New Roman') for label in labels]
plt.savefig('F:/要多写多发期刊论文啊/社区防疫 论文/做工大的吧/纯ABS要能跑出来我去死/格式冲冲冲/图重新绘制/figure10-1.png',bbox_inches='tight')
```
