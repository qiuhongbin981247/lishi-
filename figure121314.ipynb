```python
import pandas as pd
import matplotlib.pyplot as plt

data_source1 = pd.read_excel('figure11-day1.xlsx')
# 函数plot()尝试根据数字绘制出有意义的图形
#print(data_source['Susceptible(EB)'])
print(data_source1)
```

```python
figsize = 5,4
figure, ax = plt.subplots(figsize=figsize)
plt.plot(data_source1['Susceptible(Day 1)'],'-r',label='Susceptible(Day 1)',linewidth=2.0,ms=1)
plt.plot(data_source1['Exposed(Day 1)'],'b-',label='Exposed(Day 1)',linewidth=2.0,ms=1)
plt.plot(data_source1['Symptomatic(Day 1)'],'black',label='Symptomatic(Day 1)',linewidth=2.0,ms=1)
plt.plot(data_source1['Asymptomatic(Day 1)'],'gray',label='Asymptomatic(Day 1)',linewidth=2.0,ms=1)
plt.plot(data_source1['Intreated(Day 1)'],'green',label='Intreated(Day 1)',linewidth=2.0,ms=1)
plt.plot(data_source1['Recovered(Day 1)'],'gold',label='Recovered(Day 1)',linewidth=2.0,ms=1)
plt.plot(data_source1['Dead(Day 1)'],'pink',label='Dead(Day 1)',linewidth=2.0,ms=1)
plt.axis([0,120,0,21000])
# 设置x轴名称
font1 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 10,
}
plt.title('Simulation Results When Taking CTM From Day 1',font1)
plt.grid(linestyle="--")  # 设置背景网格线为虚线
#设置图例
plt.legend(loc=1, numpoints=1,prop=font1)
plt.xlabel("Days",font1)
# 设置y轴名称
plt.ylabel("The Number of Different Type Persons",font1)
#保存图片
plt.tick_params(labelsize=10)
labels = ax.get_xticklabels() + ax.get_yticklabels()
[label.set_fontname('Times New Roman') for label in labels]
plt.savefig('F:/要多写多发期刊论文啊/社区防疫 论文/做工大的吧/纯ABS要能跑出来我去死/格式冲冲冲/图重新绘制/figure12-1.png',bbox_inches='tight')
```
