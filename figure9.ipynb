```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data_source = pd.read_excel('figure9.xlsx')
print(data_source)
```

```python
#matplotlib.pyplot.bar(left, height, alpha=1, width=0.8, color=, edgecolor=, label=, lw=3)
# 绘图
figsize = 5,4
figure, ax = plt.subplots(figsize=figsize)
plt.rc('font',family='Times New Roman') #将全局字体都设置为 times new
bar_width = 0.35
plt.bar(np.arange(7),data_source['AVERAGE'],align = "center",color = "indianred",alpha = 0.6,width = bar_width)
#plt.bar(np.arange(5) + bar_width,data_source['Group2'],align = "center",color = "indianred",alpha = 0.6,width = bar_width)
# 添加y轴标签
font1 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 10,
}
plt.grid(linestyle="--")  # 设置背景网格线为虚线
plt.ylabel("The Recovered Number (Average Value)",font1)
# 设置Y轴的刻度范围
plt.ylim([2900,3050])
labels=['35','37','40','41','43','45','50']
# 添加x轴刻度标签
plt.xticks(np.arange(7),labels)
plt.xlabel("Isolation Capacity",font1)
plt.tick_params(labelsize=10)
labels = ax.get_xticklabels() + ax.get_yticklabels()
[label.set_fontname('Times New Roman') for label in labels]
# 添加标题
plt.title('The Recovered Number in Different Isolation Capacity(Hospital Capacity =20)',font1)

# 为每个条形图添加数值标签
for x,y in enumerate(data_source['AVERAGE']):
    plt.text(x,y+2,'%s' %round(y,0),ha='center')# y+50 标签的坐标

plt.savefig('F:/要多写多发期刊论文啊/社区防疫 论文/做工大的吧/纯ABS要能跑出来我去死/格式冲冲冲/图重新绘制/figure9.png',bbox_inches = 'tight')
# 显示图形
plt.show()
```
