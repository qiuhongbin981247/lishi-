```python
import pandas as pd
import matplotlib.pyplot as plt

data_source1 = pd.read_excel('figure17-symptomatic.xlsx')
# plot()
#print(data_source['Susceptible(EB)'])
#print(data_source1)
figsize = 10,6
figure, ax = plt.subplots(figsize=figsize)
plt.plot(data_source1['proportion=0.1'],'-r',label='proportion=0.1',linewidth=1.0,ms=1,linestyle='--')
plt.plot(data_source1['proportion=0.2'],'b-',label='proportion=0.2',linewidth=2.0,ms=3)
plt.plot(data_source1['proportion=0.3'],'black',label='proportion=0.3',linewidth=2.0,ms=1,linestyle='-.')
plt.plot(data_source1['proportion=0.4'],'green',label='proportion=0.4',linewidth=2.0,ms=1)
plt.plot(data_source1['proportion=0.5'],'',label='proportion=0.5',linewidth=1.0,ms=1)
plt.plot(data_source1['proportion=0.6'],'gold',label='proportion=0.6',linewidth=2.0,ms=1)
plt.plot(data_source1['proportion=0.7'],'purple',label='proportion=0.7',linewidth=2.0,ms=1)
plt.plot(data_source1['proportion=0.8'],'b-',label='proportion=0.8',linewidth=2.0,ms=1,linestyle='-.')

plt.axis([20,62,0,15000])
# x
font1 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 10,
}
plt.title('Simulation Results of the Symptomatic In Different Proportion',font1)
plt.grid(linestyle="--")  # 
#
plt.legend(loc=1, numpoints=1,prop=font1)
plt.xlabel("Days",font1)
# y
plt.ylabel("The Number of the Symptomatic",font1)
#
plt.tick_params(labelsize=10)
labels = ax.get_xticklabels() + ax.get_yticklabels()
[label.set_fontname('Times New Roman') for label in labels]
#plt.show()
plt.savefig('F:// //ABS///figure17-symptomatic.png',bbox_inches='tight')
```
